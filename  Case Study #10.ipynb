# Encoding

    The data is categorical so we’ll use LabelEncoder to convert it to ordinal.
    This approach requires the features to be of ‘category’ datatype. By default, a non-numerical column is of ‘object’ datatype. So we will have to change the type to ‘category’ before using this approach.
    

```python
df = df.astype('category')
df.dtypes
```

```python
# Using LabelEncoder to convert catergory values to ordinal
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
for column in df.columns:
    df[column] = labelencoder.fit_transform(df[column])

df.head()
```

# Feature Reduction

    veil-type has only one unique value and hence won't contribute anything to the data. So we can safely remove it.

```python
df = df.drop(["veil-type"],axis=1)
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(data=df.corr(), annot=True, linewidths=.5, cmap="magma")
plt.show()
```

    veil colour is highly correlated with gill attachment hence we can safley remove gill attachment

```python
df = df.drop(["gill-attachment"],axis=1)
```

```python
df.shape
```

    Now we have 21 columns 

```python
X = df.drop(["class"], axis=1) 
y = df["class"] 
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler  
std_scaler = StandardScaler() 

X = std_scaler.fit_transform(X)  
```

    For feature reduction pca can be used,but we will try to keep the variance high while doing so. Because if the variance value of our data falls, information loss occurs and this is undesirable. Then let's first see how much variance you have with PCA on how many features:

# PCA

```python
from sklearn.decomposition import PCA  # Import class for PCA

for i in range(2, 21):
    pca = PCA(n_components=i)
    pca.fit_transform(X)
    print("Sum of Variance Ratio in " + str(i) + " Components: ", sum(pca.explained_variance_ratio_))
```

    Hence we can maintain approximately 93% of our data by using 14 features. For this reason, we will give the number of components as 14

```python
pca = PCA(n_components=14)
pca.fit_transform(X)
print(sum(pca.explained_variance_ratio_))
```

# Modelling

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.2)
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
from sklearn.metrics import accuracy_score


```

# Logistic Regression

```python
LR = LogisticRegression()
LR_model=LR.fit(X_train, y_train)
pred=LR_model.predict(X_test)
print('The accuracy of the Logistic Regression is',accuracy_score(pred,y_test))
```

# GaussianNB

```python
GNB=GaussianNB()
GNB_model=GNB.fit(X_train , y_train)
pred=GNB_model.predict(X_test)
print('The accuracy of the NaiveBayes is',accuracy_score(pred,y_test))
```

# DecisionTreeClassifier

```python
DT=DecisionTreeClassifier()
DT_model=DT.fit(X_train, y_train)
pred=DT_model.predict(X_test)
print('The accuracy of the DecisionTreeClassifier is',accuracy_score(pred,y_test))
```

# Random Forest

```python
RF=DecisionTreeClassifier()
RF_model=RF.fit(X_train, y_train)
pred=RF_model.predict(X_test)
print('The accuracy of the RFClassifier is',accuracy_score(pred,y_test))
```
