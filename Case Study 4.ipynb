# Check whether all the zones generate the same amount of sales.


```python
#ANOVA test is used to compare more than two groups at the same time
```

```python
#H0= All zones generate the same amount of sales
#H1= All zones donot generate the same amount of sales
```

```python
F, p = stats.f_oneway(df['Zone - A'], df['Zone - A'], df['Zone - C'],df['Zone - D'])
print("p-value for significance is: ", p)
if p<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```
